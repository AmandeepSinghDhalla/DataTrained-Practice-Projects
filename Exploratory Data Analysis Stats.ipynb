# Exploratory Data Anaylsis (EDA)

EDA is the data analysis approach where we explore the main characteristics of the data and draw insights from them after the visualisation and testing.  EDA is used to find main features, the relationship between different variables, identifing which variables are important for our problems.

Various libraries like pandas, seaborn, matplootlib etc are used for EDA. We perform:

1 Read Data

2 Variable Identification, Looking the input data, what willl be the output variable.

3 Unvariate Analysis- Bar plot, Histogram plot(column exploration)

4 Bivariate analysis- Scatter plot, pair plot(more visualisation)

5 Detecting outliers/anamoly detection

6 Removing null values/ finding missing values

7 handling categorical variables

First of all we check if there is any missing value in the data. for that we use function "isnull". Then we visualise that data.

Now we check the correlation of the dataset. It will give us the relationship between different columns of the dataset.

```python
sns.heatmap(ds.corr())
```

The above plot shows us the correlation of the dataset between different columns. 

Now, if we want to put the values of the data inside the heat map, then we have an arrgument in the nss.heatmap function where by default "annot = False". annot means Annotate, so if we want the data values inside the heatmap, when switch value of annotate to true, so "annot = True".

We can also change the colour scheme using cmap.

If the figure is small then we can incresae the figure size using the syntax : plt.figure(figsize = (a,b))  before the ploting. a and b are the dimension of the figure size.

```python
plt.figure(figsize = (8,6))
sns.heatmap(ds.corr(), cmap = 'Greens', annot = True) # since annot is true, data is put into plot
```

# 3 Unvariate Analysis- Bar plot, Histogram plot(column exploration)
# Will tell us about the outliers

Now we perform the univariate analysis. In univariate analysis, we look at each coluumn seperately, we plot the box plot and look for outliers

```python
sns.boxplot(ds['fixed acidity'], color = 'green')
```

```python
sns.boxplot(ds['citric acid'])
```

We can plot each and every column one by one by writing codes or we can plot each and every column seperately using one function. As shown below

```python
clist = ds.columns.values
plt.figure(figsize = (20, 70))
for i in range(0 , len(clist)):
    plt.subplot(8,12, i + 1 )
    sns.boxplot(ds[clist[i]], color = 'yellow')
    #plt.tight_layout()
```

Outliers are present are are removed in the very first step before checking skewness. Removing the outliers may also remove the skewness from the data.

# To check skewness of Data

Now we can check the skewness of the data using distribution plot one by one.

```python
sns.distplot(ds['fixed acidity'])
```

```python
sns.distplot(ds['volatile acidity'])
```

We can also use subplot to draw the data side by side. As shown below.

```python
plt.figure(figsize = (18, 8))
for i in range(0, len(clist)):
    plt.subplot(4, 12, i + 1)
    sns.distplot(ds[clist[i]], color = 'red')
```

From the above plots, we can see that only density and ph of the data is normally distributed, rest of the features of the wine are skewed one way or another.
