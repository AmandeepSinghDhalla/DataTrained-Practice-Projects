Now we have another type of Regression Machine Learning which is called as Logistic Machine Learning. We know tah in regression learning we predict and forecast the dependent variable using the various independent variables and how th eoutput varies with the changing input variables, we are able to predict the fdifferent targets. But what if the target or the prediction is to be carried out in only 2 options aither True/False or Yes or No. Here dependent variable is for of 0 or 1

For example we have a dataset where we have to predict wheather a patient has diabetes or not. Or, we have to predict whether tomorrow it will rain or not. The output in such case is just True/False or Yes/No or in this case 0/1. Then the Linear Regression Model will not be effective in such case as it will be unable to learn all the data or the Linear Regression Model will not fit the the data properly. Suppose the data has all the values either 1 and 0, then when we draw the Linear Regression Line, then it will not cover the complete. It will learn the minimum as all ther values are either 0 or 1. In such a case, we need a model to cover all the points of the graph. So we use the Logistic Regression.

Logistic Regresssion is the rhe Regression Machine Learning which will predict in either yes or no and it will cover all the datas. The equation of the Logistig Model is just inverse or opposite of that of Linear Model. It is given by:
    
            Equation of Logistic Regression = log ( p / 1 - p )
Here p is the success rate or chances of sucess, so 1 - p are the chances of failure. Sucess means output will be yes and failure means output will be no.


The syntax for the Logistic Regression are similar to that of the Linear Regerssion and can be found in the sklearn.linear_model library. We will see it while doing Logistic Regression.

# Confusion Matrix

Similar to what we learned in the descriptive stastatics, the confusion matrix is used in the Logistic Regression. It is used to calculate various metrics in the ML. The confusion Matrics consist of two rows and two columns. The two rows are R1 - Actual Positive and R2 - Actual Negative in that order. The two columns are C1 - Predicted Positive and C2 - Predicted Negative in that order. The Confusion Matrix is as follows:

                              C1  Predicted Positive                      C2  Predicted Negative
     R1 Actual Positive             True Positive                               False Negative
     R2 Actual Negative             False Positive                              True Negative
This is the matrix which is used to calculate various metrics and error in the Logistic Regression ML.

True Positive(TP) and True Negative(TN) are the actual values which are accurate. False Negative(FN) are the values which were predicted by the model to be negative or No, but they are actually Positive. So, the model is not sure and places them in type 2 error. Similarly, False Positive(FP) are the values which were predicted by the model to be Positive or Yes, but they are actually Negative. So, the model is not sure and places them in type 1 error.

# Logistic Metrics

Now, there are some metrics or the calculation which are calculated in Logistic Regression Model. These are used to check various parametres of the ML Model.

ACCURACY - Accuracy of the model is measure of predicting correctly for the given model. Sum of True Positive(TP) and True Negative(TN) divided by the sum of all values of the confusion matrix:

        Accuracy = TP + TN / TP + FP + TN + FN
RECALL or SENSITIVITY or TRUE POSITIVE RATE - Recall or Sensitivity or TPR of the model is measure of identifing the positive instances of the given model. Given by the True Positive divided by the Actual Positive:

        Recall or Sensitivity or TPR = TP / TP + FN
FALSE NEGATIVE RATE - False Negative rate is give by:

        FNR = FN / TP + FN
PRECISION - Precision of the model is the measure of identifing only positive instances of the model. Given by True Positive divided by Predicted Positive:

        Precision = TP / TP + FP
SPECIFICITY - Specificity of the model is the measure of avoiding false alarms of the model. Given by True Negative divided by Actual Negative:

        Specificity = TN / FP + TN
F1 SCORE - F1 Score is the measure of the model performance of the Logistic Regression Model. It is given by:

        F1-Score = 2 X Precision X Recall / Precision + Recall
All the above measurements are the error measurements of the Logistic Regression Model. We donot have to calculate them manually as we have the meathod "classification_report" inside the sklearn.metrics where function for these are inbuilt.
